# Logistic Regression

Now it's time to do a train test split, and train our model! You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.model_selection import train_test_split
X = advertising_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = advertising_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
# Train and fit a logistic regression model on the training set.**
from sklearn.linear_model import LogisticRegression
# Instantiate LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations
Now predict values for the testing data.
Make adjustments to C to improve your **model**; Document multiple models to demonstrate overfitting, underfitting, and best fit.

** Create a classification report for the model, displaying the CONFUSION MATRIX**

```python
# Predictions and Evaluations
# Now predict values for the testing data.
predictions = logmodel.predict(X_test)

# Create a classification report and confusion matrix for the model.
from sklearn.metrics import classification_report, confusion_matrix
print ('Classification Report :\n ',classification_report(y_test,predictions))
print ('Confusion Matrix :\n',confusion_matrix(y_test,predictions))
# End 
```

```python
#Below graph is used for underfitting, overfitting and bestfit
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    logreg = LogisticRegression(C=C, solver='liblinear', penalty="l1").fit(X_train, y_train)
    print("Training accuracy of logreg with C={:.3f}: {:.2f}".format(
          C, logreg.score(X_train, y_train)))
    print("Test accuracy of logreg with C={:.3f}: {:.2f}".format(
          C, logreg.score(X_test, y_test)))
    plt.plot(logreg.coef_.T, marker, label="C={:.3f}".format(C))

plt.xticks(range(advertising_data.shape[1]), advertising_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']], rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.title(label="Jayarani Emekar :\n Logistic Regression model", 
          loc="left", fontsize=15,
          fontstyle='italic')
plt.ylim(-0.4, 0.2)
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
X = advertising_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = advertising_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
# Train and fit a LinearSVC model on the training set.**
from sklearn.svm import LinearSVC
# Instantiate LogisticRegression
logmodel = LinearSVC()
logmodel.fit(X_train,y_train)
```

```python
# Predictions and Evaluations
# Now predict values for the testing data.
predictions = logmodel.predict(X_test)

# Create a classification report and confusion matrix for the model.
from sklearn.metrics import classification_report, confusion_matrix
print ('Classification Report : \n',classification_report(y_test,predictions))
print ('Confusion Matrix : \n',confusion_matrix(y_test,predictions))
# End
```

```python
#below graph is used for underfitting, overfitting and bestfit
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    logsvc = LinearSVC(C=C).fit(X_train, y_train)
    print("Training accuracy of linear svc with C={:.3f}: {:.2f}".format(
          C, logsvc.score(X_train, y_train)))
    print("Test accuracy of linear svc with C={:.3f}: {:.2f}".format(
          C, logsvc.score(X_test, y_test)))
    plt.plot(logsvc.coef_.T, marker, label="C={:.3f}".format(C))

plt.xticks(range(advertising_data.shape[1]), advertising_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']], rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.title(label="Jayarani Emekar :\n Linear SVC model", 
          loc="left", fontsize=15,
          fontstyle='italic')
plt.ylim(-0.005, 0.005)
plt.legend()
```
